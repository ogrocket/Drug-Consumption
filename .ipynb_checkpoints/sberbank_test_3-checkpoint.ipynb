```python
def dict_encoding(data: pd.DataFrame, columns: list, dictionary: dict) -> pd.DataFrame:
    '''
    Function to encode df cols using your dictionary
    '''
    
    for column in columns:
        data[column] = data[column].replace(dictionary)
    return data



target = ['Alcohol', 'Amphet', 'Amyl', 'Benzo',
           'Caff', 'Cannabis', 'Choc', 'Coke', 'Crack', 'Ecstasy',
           'Heroin', 'Ketamine', 'Legalh', 'LSD', 'Meth',
           'Mushrooms', 'Nicotine', 'Semer', 'VSA']

CL = {'CL0': 0, 'CL1': 1, 'CL2': 2, 'CL3': 3, 'CL4': 4, 'CL5': 5, 'CL6': 6}

data = dict_encoding(data, target, CL)

data[target].hist(figsize=(15, 15), ec="k",
                           weights=np.ones_like(data.index)/len(data.index), color = "skyblue", bins = 15)

plt.tight_layout()
plt.show()
```

Для начала хочется свести задачу к $\textit{multilabel binary classification}$, а затем возможно и просто к $\textit{binary classification}$.

Если для данного человека и наркотика стоит одна из меток $CL0, CL1, CL2, CL3$ то скорее зависимости от данного наркотика он не имеет. Если же стоит иная, то скорее всего наоборот.

Кстати, можно заметить, что метки являются множеством с заданным порядком. Ведь если употреблял год назад, то и употреблял в последнем десятилетии, или если употреблял на последней неделе то употреблял и в последнем месяце/году/десятилетии.

Можно объединить некоторые наркотики в группы, и тем самым уменьшить количество лэйблов классификации. Мне приходит идея объединить наркотики в группы по воздействию на организм / их свойствам.

### 2. Предобработка данных.
Разделим исходную выборку на обучающую и валидационную с сохранением пропорции классов. <br>
Применим one-hot кодирование к категориальным признакам. <br>
Отмасштабируем признаки для логистической регрессии.

```python
# Train test split

Y_s = data[my_classes]

X1_train, X1_test, y1_train, y1_test = train_test_split(data[categorical + real], Y_s.iloc[:,0],
                                                        test_size=0.3, random_state=10, stratify=Y_s.iloc[:,0])
X2_train, X2_test, y2_train, y2_test = train_test_split(data[categorical + real], Y_s.iloc[:,1],
                                                        test_size=0.3, random_state=10, stratify=Y_s.iloc[:,1])
X3_train, X3_test, y3_train, y3_test = train_test_split(data[categorical + real], Y_s.iloc[:,2],
                                                        test_size=0.3, random_state=10, stratify=Y_s.iloc[:,2])
```

```python
# Label encoding

label_enc = LabelEncoder()

for ctg in categorical:
    X1_train[ctg] = label_enc.fit_transform(X1_train[ctg])
    X1_test[ctg] = label_enc.transform(X1_test[ctg])
    X2_train[ctg] = label_enc.fit_transform(X2_train[ctg])
    X2_test[ctg] = label_enc.transform(X2_test[ctg])
    X3_train[ctg] = label_enc.fit_transform(X3_train[ctg])
    X3_test[ctg] = label_enc.transform(X3_test[ctg])
```

```python
# One hot encoding

one_hot = OneHotEncoder(drop='first',sparse=False)

one_hot_df = pd.DataFrame(one_hot.fit_transform(X1_train[categorical]), index=X1_train.index) # fit_transform on train
X1_train.drop(categorical, axis=1, inplace=True)
X1_train = pd.concat((X1_train, one_hot_df), axis=1)

one_hot_df = pd.DataFrame(one_hot.transform(X1_test[categorical]), index=X1_test.index) # transform on test
X1_test.drop(categorical, axis=1, inplace=True)
X1_test = pd.concat((X1_test, one_hot_df), axis=1)



one_hot_df = pd.DataFrame(one_hot.fit_transform(X2_train[categorical]), index=X2_train.index) # fit_transform on train
X2_train.drop(categorical, axis=1, inplace=True)
X2_train = pd.concat((X2_train, one_hot_df), axis=1)

one_hot_df = pd.DataFrame(one_hot.transform(X2_test[categorical]), index=X2_test.index) # transform on test
X2_test.drop(categorical, axis=1, inplace=True)
X2_test = pd.concat((X2_test, one_hot_df), axis=1)



one_hot_df = pd.DataFrame(one_hot.fit_transform(X3_train[categorical]), index=X3_train.index) # fit_transform on train
X3_train.drop(categorical, axis=1, inplace=True)
X3_train = pd.concat((X3_train, one_hot_df), axis=1)

one_hot_df = pd.DataFrame(one_hot.transform(X3_test[categorical]), index=X3_test.index) # transform on test
X3_test.drop(categorical, axis=1, inplace=True)
X3_test = pd.concat((X3_test, one_hot_df), axis=1)
```

```python
# Scaling for logit

scaler = StandardScaler()

X1_train_lr = pd.DataFrame(scaler.fit_transform(X1_train), index=X1_train.index, columns=X1_train.columns)
X1_test_lr = pd.DataFrame(scaler.transform(X1_test), index=X1_test.index, columns=X1_test.columns)

X2_train_lr = pd.DataFrame(scaler.fit_transform(X2_train), index=X2_train.index, columns=X2_train.columns)
X2_test_lr = pd.DataFrame(scaler.transform(X2_test), index=X2_test.index, columns=X2_test.columns)

X3_train_lr = pd.DataFrame(scaler.fit_transform(X3_train), index=X3_train.index, columns=X3_train.columns)
X3_test_lr = pd.DataFrame(scaler.transform(X3_test), index=X3_test.index, columns=X3_test.columns)
```

### 3. Выбор метрики и настройка моделей.
#### Метрика
В выборках имеем сильный дисбаланс классов, значит выбор accuracy в качестве метрики будет плохим. <br>
Искать лучшую модель будем с помощь AUC-ROC, тк он говорит об обобщающей способности алгоритма. <br>
В качестве основной метрики выберем F-меру, тк она считается из precision и recall, для которых дисбаланс классов не является проблемой.

```python
stimulants_scores = []
hallucinogens_scores = []
op_and_depress_scores = []
```

#### Логистическая регрессия

```python
lr_params = { 'C': np.logspace(-5, 0, 200),
             'penalty': ['l1'],
             'solver': ['liblinear', 'saga'],
             'class_weight': ['balanced']}

logit_stimulants = GridSearchCV(LogisticRegression(), param_grid=lr_params, scoring='roc_auc')
logit_stimulants.fit(X1_train_lr, y1_train)

# set best params
logit_stimulants_b = LogisticRegression()
logit_stimulants_b.set_params(**logit_stimulants.best_params_)

logit_stimulants_b.fit(X1_train_lr, y1_train)

stimulants_scores.append(
        roc_auc_score(y1_test, logit_stimulants_b.predict_proba(X1_test_lr)[:,1])
        )
```

```python
logit_hallucinogens = GridSearchCV(LogisticRegression(), param_grid=lr_params, scoring='roc_auc')
logit_hallucinogens.fit(X2_train_lr, y2_train)

# set best params
logit_hallucinogens_b = LogisticRegression()
logit_hallucinogens_b.set_params(**logit_hallucinogens.best_params_)

logit_hallucinogens_b.fit(X2_train_lr, y2_train)


hallucinogens_scores.append(roc_auc_score(y2_test, logit_hallucinogens_b.predict_proba(X2_test_lr)[:,1]))
```

```python
logit_depressants = GridSearchCV(LogisticRegression(), param_grid=lr_params, scoring='roc_auc')
logit_depressants.fit(X3_train_lr, y3_train)

# set best params
logit_depressants_b = LogisticRegression()
logit_depressants_b.set_params(**logit_depressants.best_params_)

logit_depressants_b.fit(X3_train_lr, y3_train)

op_and_depress_scores.append(roc_auc_score(y3_test, logit_depressants_b.predict_proba(X3_test_lr)[:,1]))
```

#### Случайный лес

```python
rf_params = {'n_estimators': np.linspace(100, 1000, 10).astype(int),
               'max_features': ['sqrt'],
               'max_depth': np.linspace(1, 20, 11).astype(int),
               'min_samples_split': [2, 5, 10, 15, 20, 25, 50],
               'min_samples_leaf': [1, 2, 4, 8, 16, 32, 64]}
```

```python
rf_stimulants = RandomizedSearchCV(RandomForestClassifier(), rf_params, 
                                   n_iter=100, cv=3, scoring='roc_auc', n_jobs=-1, random_state=1)
rf_stimulants.fit(X1_train, y1_train)

# set best params
rf_stimulants_b = RandomForestClassifier()
rf_stimulants_b.set_params(**rf_stimulants.best_params_)

rf_stimulants_b.fit(X1_train, y1_train)

stimulants_scores.append(roc_auc_score(y1_test, rf_stimulants_b.predict_proba(X1_test)[:,1]))
```

```python
rf_hallucinogens = RandomizedSearchCV(RandomForestClassifier(), rf_params,
                                   n_iter=100, cv=3, scoring='roc_auc', n_jobs=-1, random_state=1)
rf_hallucinogens.fit(X2_train, y2_train)

# set best params
rf_hallucinogens_b = RandomForestClassifier()
rf_hallucinogens_b.set_params(**rf_hallucinogens.best_params_)

rf_hallucinogens_b.fit(X2_train, y2_train)

hallucinogens_scores.append(roc_auc_score(y2_test, rf_hallucinogens_b.predict_proba(X2_test)[:,1]))
```

```python
rf_depressants = RandomizedSearchCV(RandomForestClassifier(), rf_params,
                                   n_iter=100, cv=3, scoring='roc_auc', n_jobs=-1, random_state=1)
rf_depressants.fit(X3_train, y3_train)

# set best params
rf_depressants_b = RandomForestClassifier()
rf_depressants_b.set_params(**rf_depressants.best_params_)

rf_depressants_b.fit(X3_train, y3_train)

op_and_depress_scores.append(roc_auc_score(y3_test, rf_depressants_b.predict_proba(X3_test)[:,1]))
```

### 4. Сведение задачи к бинарной классификации.

В качестве итогового таргета хочется предсказывать является ли человек наркоманом или же нет. Для разметки данных воспользуемся следующей идеей: если человек относится к положительному классу хотябы для одной группы наркотиков, тогда он наркоман. Выше я уже проделал это и получил стобец 'drug_user'. Конечную модель построим следующим образом: возьмём результаты вышепостроенных моделей и обучим на них логистическую регрессию, которая будет предсказывать итоговую метку.


```python
X_train, X_test, y_train, y_test = train_test_split(data[categorical + real], Y_s.iloc[:,3],
                                                        test_size=0.3, random_state=10, stratify=Y_s.iloc[:,3])

for ctg in categorical:
    X_train[ctg] = label_enc.fit_transform(X_train[ctg])
    X_test[ctg] = label_enc.transform(X_test[ctg])

one_hot_df = pd.DataFrame(one_hot.fit_transform(X_train[categorical]), index=X_train.index) # fit_transform on train
X_train.drop(categorical, axis=1, inplace=True)
X_train = pd.concat((X_train, one_hot_df), axis=1)

one_hot_df = pd.DataFrame(one_hot.transform(X_test[categorical]), index=X_test.index) # transform on test
X_test.drop(categorical, axis=1, inplace=True)
X_test = pd.concat((X_test, one_hot_df), axis=1)

X_train_lr = pd.DataFrame(scaler.fit_transform(X_train), index=X_train.index, columns=X_train.columns)
X_test_lr = pd.DataFrame(scaler.transform(X_test), index=X_test.index, columns=X_test.columns)
```

```python
# fitting base models

logit_stimulants_b.fit(X_train_lr, y_train)
logit_hallucinogens_b.fit(X_train_lr, y_train)
logit_depressants_b.fit(X_train_lr, y_train)

rf_stimulants_b.fit(X_train, y_train)
rf_hallucinogens_b.fit(X_train, y_train)
rf_depressants_b.fit(X_train, y_train)

xgb_stimulants_b.fit(X_train, y_train);
xgb_hallucinogens_b.fit(X_train, y_train);
xgb_depressants_b.fit(X_train, y_train);
```

```python
# creating train for final model
pred1 = logit_stimulants_b.predict_proba(X_train_lr)[:,1]
pred2 = logit_hallucinogens_b.predict_proba(X_train_lr)[:,1]
pred3 = logit_depressants_b.predict_proba(X_train_lr)[:,1]

pred4 = rf_stimulants_b.predict_proba(X_train)[:,1]
pred5 = rf_hallucinogens_b.predict_proba(X_train)[:,1]
pred6 = rf_depressants_b.predict_proba(X_train)[:,1]

pred7 = xgb_stimulants_b.predict_proba(X_train)[:,1]
pred8 = xgb_hallucinogens_b.predict_proba(X_train)[:,1]
pred9 = xgb_depressants_b.predict_proba(X_train)[:,1]

X_train_preds = pd.DataFrame([pred1,pred2,pred3,pred4,pred5,pred6,pred7,pred8,pred9]).T

# creating test for final model
pred1 = logit_stimulants_b.predict_proba(X_test_lr)[:,1]
pred2 = logit_hallucinogens_b.predict_proba(X_test_lr)[:,1]
pred3 = logit_depressants_b.predict_proba(X_test_lr)[:,1]

pred4 = rf_stimulants_b.predict_proba(X_test)[:,1]
pred5 = rf_hallucinogens_b.predict_proba(X_test)[:,1]
pred6 = rf_depressants_b.predict_proba(X_test)[:,1]

pred7 = xgb_stimulants_b.predict_proba(X_test)[:,1]
pred8 = xgb_hallucinogens_b.predict_proba(X_test)[:,1]
pred9 = xgb_depressants_b.predict_proba(X_test)[:,1]

X_test_preds = pd.DataFrame([pred1,pred2,pred3,pred4,pred5,pred6,pred7,pred8,pred9]).T
```

```python
logit = GridSearchCV(LogisticRegression(), param_grid=lr_params, scoring='roc_auc')
logit.fit(X_train_preds, y_train)

# set best params
logit_b = LogisticRegression()
logit_b.set_params(**logit.best_params_)

logit_b.fit(X_train_preds, y_train)

print('AUC-ROC =',roc_auc_score(y_test, logit_b.predict_proba(X_test_preds)[:,1]))
```

### 5. Подбор порогов

```python
precision, recall, thresholds = precision_recall_curve(y_test, logit_b.predict_proba(X_test_preds)[:,1]) 
thresholds = np.append(thresholds, 1) 
predictions = logit_b.predict_proba(X_test_preds)[:,1]
```

```python
fscore = []
for threshold in thresholds:
    y_pred_prob = (predictions > threshold).astype(int)
    fscore.append(f1_score(y_test, y_pred_prob))

fscore = np.array(fscore)
```
